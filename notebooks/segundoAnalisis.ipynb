```python
import numpy as np

# Función para cargar los datos del archivo CSV
def cargar_datos(ruta_archivo):
    # Cargar los datos del archivo CSV utilizando NumPy
    datos = np.genfromtxt(ruta_archivo, delimiter=',', skip_header=1)
    return datos

if __name__ == "__main__":
    # Ruta del archivo CSV
    ruta_archivo = '../retail_sales_dataset.csv'
    
    # Cargar los datos
    datos = cargar_datos(ruta_archivo)
    
   
```

```python
import pandas as pd

"""Muestra las primeras 10 filas del DataFrame para confirmar que los datos se han cargado correctamente."""
df = pd.DataFrame(datos)

print(df.head(10))
```

```python
"""Muestra las últimas 5 filas del DataFrame."""
import pandas as pd

df = pd.DataFrame(datos)
print("\nÚltimas filas del DataFrame:")
print(df.tail(5))

"""Utiliza el método info() para obtener información general sobre el DataFrame, incluyendo el número de entradas, nombres de las columnas, tipos de datos y memoria utilizada."""
print("\nInformación del DataFrame:")
df = pd.read_csv('../retail_sales_dataset.csv', header=0)
print(df.info())

"""Genera estadísticas descriptivas del DataFrame utilizando el método describe()."""
print("\nEstadísticas descriptivas del DataFrame:")
df.describe()

"""Inspecciona los tipos de datos de cada columna utilizando el atributo dtypes."""
print("\nTipos de datos de las columnas:")
print(df.dtypes)

"""Cuenta los valores únicos en la columna Producto utilizando el método value_counts()."""
print("\nConteo de valores únicos en la columna 'Product Category':")
print(df['Product Category'].value_counts())

"""Muestra todos los valores únicos en la columna Tienda utilizando el método unique()."""
print("\nValores únicos en la columna 'Gender':")
print(df['Gender'].unique())

"""Filtra el DataFrame para mostrar solo las filas donde las ventas (Ventas) sean mayores a 50."""
totalAmount_mayores_50 = df[df['Total Amount'] > 50]
print("\nFilas donde los valores deTotalAmount son mayores a 50:")
print(totalAmount_mayores_50)

"""Filtra el DataFrame para mostrar solo las filas donde el precio (Precio) sea menor a 0.5."""
priceperUnit_05 = df[df['Price per Unit'] < 30]
print("\nFilas donde los valores de Price per Unit son menores a 30:")
print(priceperUnit_05)

"""Utilizando el método query(), filtra el DataFrame para mostrar las filas donde el producto sea Manzanas y las ventas sean mayores a 30."""
query_result = df.query('`Total Amount` > 50 & `Product Category` == "Clothing"')
print("\nFilas donde el producto sea Clothing y el totalAmount sean mayores a 50:")
print(query_result)

"""Selecciona y muestra solo las columnas Producto y Ventas del DataFrame."""

sliced_columns = df[['Product Category', 'Total Amount']]
print("\n Columnas Product Category y Total Amount")
print(sliced_columns)

"""Utilizando loc[], selecciona y muestra las filas de la 5 a la 10 (inclusive) y las columnas Producto y Tienda."""

loc_result = df.loc[5:10, ['Product Category', 'Total Amount']]
print("\n Filas desde la 5 a la 10 de Product category y Total Amoount ")
print(loc_result)

"""Utilizando iloc[], selecciona y muestra las primeras 5 filas y las primeras 3 columnas del DataFrame."""

iloc_result = df.iloc[0:5, 0:3]
print("\n Primeras 5 filas y las primeras 3 columnas del DataFrame ")
print(iloc_result)
```
